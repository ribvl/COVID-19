```python
def read_csv_columns(path, file_name):
    file_columns = list(pd.read_csv(path+"\\"+file_name, sep=';', nrows=2).columns)
    return file_columns

def read_csv_data(path, file_name):
    files = pd.read_csv(path+"\\"+file_name, sep=';')
    #files = files.T
    return files
#     size_limit = 25
#     if os.path.getsize(path+"\\"+file_name)/1000000 < size_limit:
#         files = pd.read_csv(path+"\\"+file_name, sep=';')
#         # files = files.T
#         return files
```

```python
def file_columns(path, list_of_files):
    df = pd.DataFrame({"None": "None"}.items(), columns=['file', 'columns'])
    
    for file in list_of_files:
        new_row = pd.DataFrame({file: '; '.join(str(i) for i in read_csv_columns(path,file))}.items(), columns=['file', 'columns'])
        df = df.append(new_row, ignore_index=True)
    
    df = df.iloc[1:]

    return df

file_columns = file_columns(path, csv_files)
file_columns.head()
```
